# 4. On average which region has the highest 
a. GDP per capita (current US$) b. International trade: Imports (million US$)

```python
M1=data[['Region','GDP per capita (current US$)']].groupby(['Region']).agg('mean')
M1
```

```python
M1[M1['GDP per capita (current US$)']==M1['GDP per capita (current US$)'].max()]
```

```python
#North America has the highest GDP average
```

```python
M2=data[['Region','International trade: Imports (million US$)']].groupby(['Region']).agg('mean')
M2
```

```python
M2[M2['International trade: Imports (million US$)']==M2['International trade: Imports (million US$)'].max()]
```

```python
#North America has the highest International trade average
```
