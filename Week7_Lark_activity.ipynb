# 2. Check whether female and male students scored the same marks.

```python
from scipy.stats import ttest_ind
import numpy as np
male=data[data['Gender']=='male']['Total Marks']
female=data[data['Gender']=='female']['Total Marks']
ttest,pval = ttest_ind(male,female)
print("p-value=",pval,",","ttest value=",ttest)
if pval <0.05:
  print("We reject null hypothesis. Female and male students  scored the same marks")
else:
  print("We accept null hypothesis that female and male students didn't score the same marks")
```

# 3. Whether test preparation helps the students?

```python
from scipy.stats import ttest_ind
import numpy as np
notest=data[data['Test preparation']=='none']['Total Marks']
donetest=data[data['Test preparation']=='completed']['Total Marks']
ttest,pval = ttest_ind(notest,donetest)
print("p-value=",pval,",","ttest value=",ttest)
if pval <0.05:
  print("We reject null hypothesis. Test preparation helps the students.")
else:
  print("We accept null hypothesis that test preparation doesn't help the students.")
```
