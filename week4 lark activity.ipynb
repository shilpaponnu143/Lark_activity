```python
import pandas as pd
data=pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data
```

# 1.Compare churn count with respect to gender


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.countplot(x=data['gender'],hue=data['Churn'],data=data)
plt.xlabel("Gender",fontsize=16)
plt.ylabel("Churn count",fontsize=16)
plt.title("Gender vs Churn count",fontsize=16)
plt.grid(color = "grey", linewidth = "1.4", linestyle = "--")
plt.show()
```

# .2. Find out how many female senior citizens there in the dataset

```python
sns.countplot(x=F['SeniorCitizen'],data=F)
plt.xlabel(' NO:OF SENIOR CITIZENS',fontsize=16)
plt.ylabel("count",fontsize=16)
plt.title('PLOT OF FEMALE SENIOR CITIZENS',fontsize=16)
plt.grid(color = "grey", linewidth = "1.4", linestyle = "--")
plt.show()
```

# 3. Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(6,4))
plt.scatter(data['tenure'],data['TotalCharges'],s=10)
plt.title('plot of Tenure vs Totalcharges',fontsize=16)
plt.xlabel('TENURE')
plt.ylabel('TOTALCHARGES')
```

# 4. Find out which contract is preferred by the senior citizen

```python
F=data[data['SeniorCitizen']==1]
F
```

```python
sns.countplot(x=F['Contract'], data=F)
plt.xlabel("Contract",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title("Distribution of contract preferred by Senior citizens",fontsize=15)
plt.grid(color = "grey", linewidth = "1.4", linestyle = "--")
```

```python
# Month-to-month contract is preferred by senior citizens
```
