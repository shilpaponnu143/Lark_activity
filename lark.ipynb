```python
import pandas as pd
```

```python
data=pd.read_csv('company_sales_data.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
F=data[(data['facewash']<1500)]
```

```python
F
```

```python
import numpy as np
barWidth = 0.25
x=F['month']
br1 = np.arange(len(F))
br1
br2 = br1+barWidth
Height1=F['facewash']
Height2=F['facecream']
plt.bar(br1,Height1,color='red',width = barWidth,label ='face Wash sales')
plt.bar(br2,Height2,color='green',width = barWidth,label ='face Cream sales')
plt.xlabel("Month")
plt.ylabel("Sales units in number")
plt.xticks((br1+br2)/2,x)
plt.legend()
plt.grid(color = "grey", linewidth = "1.4", linestyle = "--")
plt.figure(1).text(0.5, 0.95 , "Facewash Sales vs Facecream Sales\n", ha="center", va="bottom", size="large",fontweight='bold')
plt.figure(1).text(0.5,0.95,"For the months facewash sales less than 1500", ha="center", va="bottom", size="medium")
#plt.show()
```
